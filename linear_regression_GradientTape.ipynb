#  Gradient Tapes

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# cearte data points 
slope = 0.6
mu = 0
sigma = 5
noise = np.random.normal(mu,sigma,100)

x_data = np.arange(0,100)
trend = slope*x_data +5
y_data = trend+noise

plt.plot(x_data,trend)
plt.plot(x_data,y_data,'o')
```

```python
# least square method
n = x_data.size
m_least_square_method = (n*sum(x_data*y_data)-sum(x_data)*sum(y_data))/(n*sum(x_data*x_data)-(sum(x_data))**2) 
constant_least_square_method = (sum(y_data)- m_least_square_method *sum(x_data)) / n
print('slope =', m_least_square_method)
print('constant =', constant_least_square_method)
```

```python
plt.plot(x_data,trend,'g')
plt.plot(x_data,y_data,'bo')
prediction_least_square_method = m_least_square_method*x_data + constant_least_square_method
plt.plot(x_data,prediction_least_square_method,'--r')
```

```python
# Gradient tapes

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# cearte data points 
slope = 2
constant = 5
mu = 0
sigma = 5
noise = np.random.normal(mu,sigma,100)

x_data = np.arange(0,100)
trend = slope*x_data + constant
y_data = trend+noise
n = x_data.size

plt.plot(x_data,trend)
plt.plot(x_data,y_data,'o')


W = tf.Variable(np.random.rand())
b = tf.Variable(np.random.rand())
lr = 2*1e-4

print(0, W.numpy(), b.numpy())   
for epoch in range(100):
    with tf.GradientTape(persistent =True) as t:
        y = W*x_data + b
        loss = sum((y-y_data)**2)/n
    dW = t.gradient(loss,W)
    db = t.gradient(loss,b)
    b.assign_sub(5e1*lr*db)
    W.assign_sub(lr*dW)
    print(epoch+1, W.numpy(), b.numpy(), loss.numpy())   
```
